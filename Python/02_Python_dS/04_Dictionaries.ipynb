## Question 4
Count character frequencies in a word using a dictionary.


```python
word = "banana"
freq = {}

for char in word:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print(freq)
```

## Question 5
Create a dictionary comprehension that maps numbers 1-5 to their squares.

```python
squares = {num: num**2 for num in range(1, 6)}

print(squares)
```
