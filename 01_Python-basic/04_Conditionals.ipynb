## Question 1
Write a program that checks whether a number is positive, negative, or zero.

```python
num = int(input("Enter a number: "))
if num==0:
    print(f"{num} is 0")
elif num>0:
    print(f"{num} is positive")
else:
    print(f"{num} is negetive")
```

## Question 3
Take marks as input and print Grade A/B/C/D/F using if-elif-else.

```python
mark=int(input("enter ur marks: "))
if mark>=90:
    print("A grade")
elif mark>=70:
    print('B grade')
elif mark>=50:
    print('C grade')
elif mark>=30:
    print('D grade')
else:
    print("f grade")
```

## Question 4
Check whether a year is a leap year.

```python
year =int(input('enter year'))
if year%4 == 0:
    print(f'this year {year} a leap year')
else:
    print('not a leap year')
```

## Question 5
Write a one-line conditional expression (ternary operator) to find the larger of two numbers.

```python
num1=int(input('enter 1st num'))
num2=int(input('enter 2nd num'))

max = num1 if num1>num2 else num2
print(f"the max is {max}")
```
