## Question 1
Print numbers from 1 to 20 using a loop.

```python
for x in range(20):
    print(x+1)
```

## Question 2
Print all even numbers between 1 and 30.

```python
for x in range(1,31):
    if x%2 == 0:
        print(x)
    else:
        pass
```

## Question 3
Find the sum of numbers from 1 to 100.

```python
sum:int=0
for x in range(101):
    sum+=x
print(sum)
```

## Question 4
Print the multiplication table of a given number.

```python
num=int(input("enter a num: "))
for x in range(1,11):
    print(f"{num} x {x} = {num*x}")
```

## Question 5
Count how many vowels are present in the string 'Artificial Intelligence'.

```python
count:int = 0
vowel = ["A","a","E","e","I","i","O","o","U","u"]
ai:str = 'Artificial Intelligence'
for x in ai:
    if x in vowel:
        count+=1
    else:
        pass
count
```
