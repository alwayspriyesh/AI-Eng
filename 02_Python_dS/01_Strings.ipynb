```python
string3:str="yo soy priyesh"
reversed_string3:str = ""
for x in range(len(string3)-1, -1, -1):
    reversed_string3+= string3[x]

reversed_string3
```

## Question 5
Check whether a string is a palindrome.

```python
str4:str = "madam"
for x in range(len(str4)//2):
    if str4[x] != str4[len(str4) - 1 - x]:
        is_palind:bool = False
        break
else:
    is_palind:bool = True

print(is_palind)
```
